# Divisor

```python
numero = int(input("Número: "))
divisor = int(input("Divisor: "))
count = 0

# Número positivo:
# range laço: (inicio, fim, passo)


for i in range(0, abs(numero), abs(divisor)):
    count = count + 1
    
if(numero > 0):
    if(numero - abs(divisor*count) < 0):
            count = count - 1
    if(divisor < 0):
        count = count * (-1)
        
if(numero < 0):
    if(abs(numero) - abs(divisor*count) < 0):
            count = count - 1
    if(divisor > 0):
        count = count * (-1)
        
mod = numero - (divisor * count)
print("Resultado: ", count)
print("Resto: ", mod)
```
